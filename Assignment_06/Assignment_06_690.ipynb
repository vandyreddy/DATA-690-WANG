Clean Census Year

```python
def cleanse_year(string):
    string = string.replace("*","")
    return string
```

```python
cleanse_year("1989*")
```

```python
cleanse_year("20*10")
```

```python
us_info["new census year"] = us_info["Census Year"].apply(cleanse_year)
us_info
```

Clean Census Cost 

```python
def census_cost(input_cost):
    cost = input_cost.lstrip("$").replace(",","")
    new_cost=cost.replace('Billion','0000000000')
    return new_cost
```

```python
us_info["new cost"] = us_info["Census Cost"].apply(census_cost)
us_info
```

Clean Average Cost

```python
def avgCost(avg_cost):
  avg_costs = avg_cost.replace('$','')
  if 'cents' in avg_costs:
    avg_costs = avg_costs.replace('cents','')
    avg_costs = float(avg_costs)/100
  return str(avg_costs)
```

```python
us_info["new avg cost"]=us_info["Average Cost Per Person"].apply(avgCost)
us_info
```

```python
us_info.to_csv("census_cost_cleansed.csv",index=False)
```

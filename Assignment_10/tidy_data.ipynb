# DATAFRAME-2

```python
#Moving on to the second dataframe 
pew=pd.read_csv('pew.csv')
pew.head()
```

```python
#Melt is used to make the long data to the wide data 
#Here column headers are values,not variables
#So,by melting we can get correct data frame 
pew_melt=pd.melt(pew,id_vars='religion')
pew_melt.head()
```

```python
#id_vars= to use as identifier variables
#var_name=Name to use for the ‘variable’ column. If None it uses frame.columns.name or ‘variable’.
#value_name=Name to use for the ‘value’ column.
#changing headears names 
pew_melt_long=pd.melt(pew,id_vars='religion',var_name='income',value_name='count')
pew_melt_long.head()
```

```python
#Importing country_series csv 
ebola=pd.read_csv('country_timeseries.csv')
ebola.head(10)
```

```python
ebola.head(3)
```

```python
#In above data frame varibles are stored as both rows and columns 
#By specifying id_vars,which need to be specified that which columns are to be included 
ebola_long=pd.melt(ebola,id_vars=['Date','Day'])
ebola_long.head()
```

```python
#Splitting the variable 
'Cases_Guinea'.split('_')
```

```python
#Spliting the varibale in a dataframe 
variable_split=ebola_long['variable'].str.split('_')
variable_split
```

```python
#Which tells that ,dataframe is a bunch of series
#Series a bunk of 1-dimensional values 
type(variable_split)
```

```python
#Since the variable has been splitted 
#so here get(0) means the first element and get(1) is the second element 
variable_split.str.get(0)
```

```python
#Creating the two columns named stats and country 
#Stats column contains the first element that is get(0)
#Country contains the second element that is get(1)
ebola_long['stats']=variable_split.str.get(0)
ebola_long['country']=variable_split.str.get(1)
```

```python
#Now we are just expanding with the _ varible,but not splitting 
s=ebola_long['variable'].str.split('-',expand=True)
s.head()
```

```python
#Creating the stat-e,country-e columns,and then merging into the elong dataframe 
ebola_long[['stats-e','country-e']]=(ebola_long['variable'].str.split('_',expand=True))
ebola_long.head()
```
